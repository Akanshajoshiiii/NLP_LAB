## Normalization for Faster Neural Network Training

### What is Normalization?
Pixel values are scaled from
0-255 to 0-1



### Why is it Important?
-  **Faster convergence**
- **Stable gradients**
- **Better accuracy**




```python
x_train=x_train.reshape(-1,28*28)/255.0
x_test=x_test.reshape(-1,28*28)/255.0
# x_train=x_train/255
#also flattened
#x_train.reshape(-1,784)
```

```python
x_train[0] #values normalised
```

**METHOD -1**

```python
# What it does
# Uses list-style Sequential API
# Input is already flattened (784)
#flattening done before model

model=Sequential(
    [
        #list of layers jis order m krna
        #dense layer=fully conn layer, every neuron of prev layer connected to nxt layer

        keras.Input(shape=(784,)),
        layers.Dense(512,activation='relu'),
        layers.Dense(256,activation='relu'),
        layers.Dense(10,activation='softmax'),

    ]
)
```

```python
#now we specify input shape in the first dense layer
model = Sequential()
model.add(layers.Dense(512, activation='relu', input_shape=(784,)))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
# metrics=['accuracy'] means “Show me how accurate my model is during training and validation.”
model.compile(optimizer=keras.optimizers.Adam(),
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])
```

Used **SparseCategoricalCrossentropy**()  as usme labels ko 1-hot encode krne ki jarurat nhi pdti  
**CategoricalCrossentropy** me we need to one-hot encode the labels

```python
model.fit(x_train,y_train,epochs=10,batch_size=64,verbose=1,validation_data=(x_test,y_test))
```

```python
# Functional API
inputs = keras.Input(shape=(784,))

x = layers.Dense(512, activation='relu')(inputs)
x = layers.Dense(256, activation='relu')(x)

outputs = layers.Dense(10)(x)  # no softmax → logits

model1 = keras.Model(inputs=inputs, outputs=outputs)


```

```python
model1.compile(
    optimizer=keras.optimizers.Adam(),
              loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```


### Logits
**Logits are the raw output scores** produced by the last layer of a neural network **before** applying Softmax.  
They can be any real numbers (positive, negative, or zero).

Example:Logits = [2.3, -1.2, 0.5]  
These are **not probabilities**.

---

### Softmax
**Softmax converts logits into probabilities** between 0 and 1, and their sum becomes 1.

Example:
Probabilities = [0.75, 0.05, 0.20]

**What logits=True does?**
The model outputs logits, so apply Softmax internally before computing the loss.”




`verbose` controls **how much training information is printed on the screen**.



#### `verbose = 0`
- No output
- Training runs silently

#### `verbose = 1`
- Progress bar
- Shows batch-level updates (default)


#### `verbose = 2`
- One line **per epoch**
- No progress bar (clean output)


```python
model1.fit(x_train,y_train,epochs=10,batch_size=32,verbose=2,validation_data=(x_test,y_test))
```
